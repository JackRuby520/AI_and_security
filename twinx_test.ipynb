```
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
plt.style.use('ggplot')
```

```
timestamp = pd.date_range('2018-01-01', periods=100, freq='D')
y1 = 5*np.random.random_sample((100,))
y2 = 5*np.random.random_sample((100,))-2
```

```
fig = plt.figure(figsize=(10,5))

ax1 = fig.add_subplot(111)
ax1.plot(timestamp, y1)
plt.xticks(rotation=90)


ax2 = ax1.twinx()
ax2.plot(timestamp, y2, color='blue')
```

```
fig, ax1 = plt.subplots(figsize=(10,5))

sns.lineplot(timestamp, y1, ax=ax1)
plt.xticks(rotation=90)

ax2 = ax1.twinx()
sns.lineplot(timestamp, y2, ax=ax2, color='g')
```
